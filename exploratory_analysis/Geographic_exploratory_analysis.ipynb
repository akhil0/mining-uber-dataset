## Commands used for parsing the raw uber dataset
#### Command to fetch the float values of lat, long, number of pick ups per lat - long 
awk -F "," '{printf ("%.4f,%.4f\n", $2,$2, 3) }' uber-raw-data-apr14.csv | awk -F "," '{a[$0]++} END{for(i in a) print i","a[i]}'

```python
import pandas as pd
# Read the parsed data
dataframe_april = pd.read_csv("april.csv", header=0, sep=",", index_col=0)
dataframe_may = pd.read_csv("may.csv", header=0, sep=",", index_col=0)
dataframe_june = pd.read_csv("june.csv", header=0, sep=",", index_col=0)
dataframe_july = pd.read_csv("july.csv", header=0, sep=",", index_col=0)
dataframe_august = pd.read_csv("august.csv", header=0, sep=",", index_col=0)
dataframe_september = pd.read_csv("september.csv", header=0, sep=",", index_col=0)
```

## Month of April 

```python
import matplotlib  
import matplotlib.pyplot as plt  

def plotNYMap(df): 
    pd.options.display.mpl_style = 'default' #Better Styling  
    new_style = {'grid': False} #Remove grid  
    matplotlib.rc('axes', **new_style)  
    from matplotlib import rcParams  
    rcParams['axes.facecolor'] = 'black' #Size of figure  
    rcParams['figure.figsize'] = (17.5, 17) #Size of figure  
    rcParams['figure.dpi'] = 250
    P=df.plot(kind='scatter', x='pickup_longitude', y='pickup_latitude',color='black',xlim=(-74.06,-73.77),ylim=(40.61, 40.91),s=.02,alpha=1)

print(dataframe_april.describe())
plotNYMap(dataframe_april)
```
