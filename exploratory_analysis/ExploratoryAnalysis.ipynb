```python
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt

path =r'./' # use your path
all_files = glob.glob(path + "/*.csv")
df = pd.concat((pd.read_csv(f) for f in all_files))
```

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'], format="%m/%d/%Y %H:%M:%S")
df['MonthDayNum'] = df['Date/Time'].dt.day
df['HourOfDay'] = df['Date/Time'].dt.hour
df['MonthOfYear'] = df['Date/Time'].dt.month
```

```python
df_MonthDay = df.pivot_table(index=['MonthDayNum'],
                                  values='Base',
                                  aggfunc='count')
df_MonthDay = df_MonthDay/6;
df_MonthDay.plot(kind='bar', figsize=(8,6))
plt.ylabel('Average Pickups')
plt.title('Average Pickups by Day');
```

```python
df_HourDay = df.pivot_table(index=['HourOfDay'],
                                  values='Base',
                                  aggfunc='count')
df_HourDay =  df_HourDay/180
df_HourDay.plot(kind='bar', figsize=(8,6))
plt.ylabel('Average Pickups')
plt.title('Average Pickups by Hour');
```

```python
df_MonthDay = df.pivot_table(index=['MonthOfYear'],
                                  values='Base',
                                  aggfunc='count')
#df_MonthDay = df_MonthDay/6;
df_MonthDay.plot(kind='bar', figsize=(8,6))
plt.ylabel('Average Pickups')
plt.title('Average Pickups by Month');
```
